```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('C:\\Users\\Manogna\\OneDrive\\Desktop\\prodigy\\house price prediction\\train.csv')
```

```python
ds
```

```python
features = ["GrLivArea", "BedroomAbvGr", "FullBath"]
```

```python
X = ds[features]
```

```python
Y = ds["SalePrice"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred)
```

```python
r2 = r2_score(Y_test, Y_pred)
```

```python
print(f'Mean Squaed Error: {mse}')
print(f'R-Squared: {r2}')
```

```python
# plotting predictions aganist actual values
plt.scatter(Y_test, Y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
new_data = pd.DataFrame({'GrLivArea': [2500], 'BedroomAbvGr': [3], 'FullBath': [2]})
predicted_price = model.predict(new_data)
print(f'Predicted Price: {predicted_price[0]}')
```
